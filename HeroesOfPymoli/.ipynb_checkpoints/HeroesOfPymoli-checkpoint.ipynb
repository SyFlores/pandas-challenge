```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas dataframe
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data['SN'].unique())

player_count_df = pd.DataFrame([{"Total Players" : player_count}])
player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items_unique = len(purchase_data['Item ID'].unique())

item_avg_price = purchase_data['Price'].mean()

item_purchases = len(purchase_data['Purchase ID'])

item_revenue = item_avg_price * item_purchases

purchasing_analysis_df = pd.DataFrame([{
    "Number of Unique Items" : items_unique,
    "Average Price" : round(item_avg_price, 2),
    "Number of Purchases" : item_purchases,
    "Total Revenue" : item_revenue
}])

purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map('${:,.2f}'.format)

purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_sn_unique = purchase_data.drop_duplicates(['SN', 'Gender'])

gender_sn_count = gender_sn_unique['Gender'].value_counts()
gender_sn_count_df = pd.DataFrame(gender_sn_count)

gender_sn_percent = gender_sn_unique['Gender'].value_counts('SN')
gender_sn_percent_df = pd.DataFrame(gender_sn_percent)

gender_sn_count_df['Percentage of Players'] = gender_sn_percent_df['Gender']

gender_demographics_df = gender_sn_count_df.rename(columns = {'Gender' : 'Total Count'})

gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:.2%}".format)

gender_demographics_df
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_groupby = purchase_data.groupby("Gender")
gender_person_groupby = purchase_data.groupby(["SN", "Gender"])

gender_purchasecount = gender_groupby["Purchase ID"].count()
gender_purchasecount_df = pd.DataFrame(gender_purchasecount)

gender_avgprice = gender_groupby["Price"].mean()
gender_avgprice_df = pd.DataFrame(gender_avgprice)

gender_totalprice = gender_groupby["Price"].sum()
gender_totalprice_df = pd.DataFrame(gender_totalprice)

gender_person_avgprice_total = gender_person_groupby["Price"].sum()
gender_person_avgprice_total_df = pd.DataFrame(gender_person_avgprice_total) 

new_gender_person_groupby = gender_person_avgprice_total_df.groupby("Gender").mean()
new_gender_person_groupby_df = pd.DataFrame(new_gender_person_groupby)

purchasing_analysis_gender_df = gender_purchasecount_df
purchasing_analysis_gender_df = purchasing_analysis_gender_df.rename(columns = {'Purchase ID' : 'Purchase Count'})
purchasing_analysis_gender_df["Average Purchase Price"] = round(gender_avgprice_df, 2)
purchasing_analysis_gender_df["Total Purchase Value"] = gender_totalprice_df
purchasing_analysis_gender_df["Avg Total Purchase per Person"] = round(new_gender_person_groupby_df, 2)

purchasing_analysis_gender_df["Average Purchase Price"] = purchasing_analysis_gender_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = purchasing_analysis_gender_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_gender_df["Avg Total Purchase per Person"] = purchasing_analysis_gender_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

purchasing_analysis_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()



* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Binning
age_list = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 1000]

age_data = gender_sn_unique

age_data["Age Ranges"] = pd.cut(age_data["Age"], bins = age_bins, labels = age_list, right = False, include_lowest = True)


# Grouping

# age_groupby = age_data.groupby("Age Group")
# age_count = pd.DataFrame(age_groupby["SN"].count())
# age_count

age_count = age_data['Age Ranges'].value_counts()
age_count_df = pd.DataFrame(age_count)

age_percent = age_data['Age Ranges'].value_counts('SN')
age_percent_df = pd.DataFrame(age_percent)

age_count_df['Percentage of Players'] = round(age_percent_df['Age Ranges'], 4)

age_demographics_df = age_count_df.rename(columns = {'Age Ranges' : 'Total Count'})
age_demographics_df.sort_index(inplace = True)

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2%}".format)

age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_data_full = purchase_data
age_data_full["Age Ranges"] = pd.cut(age_data_full["Age"], bins = age_bins, labels = age_list, right = False, include_lowest = True)

age_groupby = age_data_full.groupby("Age Ranges")
age_person_groupby = age_data_full.groupby(["SN", "Age Ranges"])

age_purchasecount = age_groupby["Purchase ID"].count()
age_purchasecount_df = pd.DataFrame(age_purchasecount)

age_avgprice = age_groupby["Price"].mean()
age_avgprice_df = pd.DataFrame(age_avgprice)

age_totalprice = age_groupby["Price"].sum()
age_totalprice_df = pd.DataFrame(age_totalprice)

age_person_avgprice_total = age_person_groupby["Price"].sum()
age_person_avgprice_total_df = pd.DataFrame(age_person_avgprice_total)

new_age_person_groupby = age_person_avgprice_total_df.groupby("Age Ranges").mean()
new_age_person_groupby_df = pd.DataFrame(new_age_person_groupby)

purchasing_analysis_age_df = age_purchasecount_df
purchasing_analysis_age_df = purchasing_analysis_age_df.rename(columns = {'Purchase ID' : 'Purchase Count'})
purchasing_analysis_age_df["Average Purchase Price"] = round(age_avgprice_df, 2)
purchasing_analysis_age_df["Total Purchase Value"] = age_totalprice_df
purchasing_analysis_age_df["Avg Total Purchase per Person"] = round(new_age_person_groupby_df, 2)

purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_age_df["Total Purchase Value"] = purchasing_analysis_age_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_age_df["Avg Total Purchase per Person"] = purchasing_analysis_age_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

purchasing_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_groupby = purchase_data.groupby("SN")

sn_count = sn_groupby["SN"].count()
sn_count_df = pd.DataFrame(sn_count)

sn_avgprice = sn_groupby["Price"].mean()
sn_avgprice_df = pd.DataFrame(sn_avgprice)

sn_total = sn_groupby["Price"].sum()
sn_total_df = pd.DataFrame(sn_total)

top_spenders_df = sn_count_df
top_spenders_df = top_spenders_df.rename(columns = {'SN' : 'Purchase Count'})
top_spenders_df["Average Purchase Price"] = round(sn_avgprice_df, 2)
top_spenders_df["Total Purchase Value"] = sn_total_df

top_spenders_df = top_spenders_df.nlargest(5, "Total Purchase Value")

top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_item_groupby = purchase_data.groupby(["Item ID", "Item Name"])

pop_item_count = pop_item_groupby["Item ID"].count()
pop_item_count_df = pd.DataFrame(pop_item_count)

pop_item_price = pop_item_groupby["Price"].mean()
pop_item_price_df = pd.DataFrame(pop_item_price)

pop_item_total = pop_item_groupby["Price"].sum()
pop_item_total_df = pd.DataFrame(pop_item_total)

popular_items_df = pop_item_count_df
popular_items_df = popular_items_df.rename(columns = {'Item ID' : 'Purchase Count'})
popular_items_df["Item Price"] = round(pop_item_price_df, 2)
popular_items_df["Total Purchase Value"] = pop_item_total_df

profitable_items_df = popular_items_df

popular_items_df = popular_items_df.sort_values(by = ["Purchase Count"], ascending = False)

popular_items_df = popular_items_df.nlargest(5, "Purchase Count")

popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:,.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:,.2f}".format)

popular_items_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitable_items_df = profitable_items_df.nlargest(5, "Total Purchase Value")

profitable_items_df["Item Price"] = profitable_items_df["Item Price"].map("${:,.2f}".format)
profitable_items_df["Total Purchase Value"] = profitable_items_df["Total Purchase Value"].map("${:,.2f}".format)

profitable_items_df
```
